```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
city = pd.read_csv('city_data.csv')
world = pd.read_csv('global_data.csv')
city.set_index('year', inplace=True)
world.set_index('year', inplace=True)
city.head(5)
```

```python
world.head(5)
```

```python
sns.set()
world['MA_15'] = world['avg_temp'].rolling(15).mean()
world['MA_25'] = world['avg_temp'].rolling(25).mean()
world['MA_35'] = world['avg_temp'].rolling(35).mean()
world['MA_15'].plot(figsize=(14,8))
world['MA_25'].plot(figsize=(14,8))
world['MA_35'].plot(figsize=(14,8))
plt.legend()
plt.ylabel('Temperature(Celsius)')
plt.xlabel('Year')
plt.show()
```

```python
city['MA_15'] = city['avg_temp'].rolling(15).mean()
city['MA_25'] = city['avg_temp'].rolling(25).mean()
city['MA_35'] = city['avg_temp'].rolling(35).mean()
city['MA_15'].plot(figsize=(14,8))
city['MA_25'].plot(figsize=(14,8))
city['MA_35'].plot(figsize=(14,8))
plt.legend()
plt.ylabel('Temperature(Celsius)')
plt.xlabel('Year')
plt.show()
```

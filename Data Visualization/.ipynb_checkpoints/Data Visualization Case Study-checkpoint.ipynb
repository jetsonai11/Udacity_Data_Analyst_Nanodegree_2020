# Data Visualization Case Study: Analyzing a Diamond Dataset
## Dataset Overview

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

%matplotlib inline
```

```python
df = pd.read_csv('data/diamonds.csv')
df.head(3)
```

```python
# high-level overview of data shape and composition
print(df.shape)
print(df.dtypes)
df.head(10)
```

```python
# descriptive statistics for numeric variables
df.describe()
```

## Univariate Exploration

Let's start our exploration by looking at the main variable of interest: price. Is the distribution skewed or symmetric? Is it unimodal or multimodal?

```python
plt.figure(figsize=[10,6])
plt.hist(data=df, x='price', bins=100)
plt.xlabel('Price')
```

> Looks highly skewed, a log transform is warranted.

```python
log_data = np.log10(df['price'])
log_data.describe()
```

```python
plt.figure(figsize=[20,6])

plt.subplot(1,2,1)
plt.hist(data=df, x='price', bins=100)

# log trasnform applied
plt.subplot(1,2,2)
bin_edges=10**np.arange(log_data.min(), log_data.max()+0.05, 0.05)
plt.hist(data=df, x='price', bins=bin_edges);
plt.xscale('log')
xticks=[30, 100, 300, 1000, 3000, 10000, 30000, 100000]
plt.xticks(xticks, xticks)
```

Next, create a plot of our first diamond 'C' metric: carat weight. Is there an interesting pattern in these values?

```python
df['carat'].describe()
```

```python
bin_edges = np.arange(0.2, 5+0.05, 0.05)
```

```python
plt.figure(figsize=[10,6])
plt.hist(data=df, x='carat', bins=bin_edges)
plt.xlabel('Weight')
plt.xlim(0, 3)
```

Now, let's move on to exploring the other three 'C' quality measures: cut, color, and clarity. For each of these measures, does the data we have tend to be higher on the quality scale, or lower?
> **Remember to always order the data for ordinal data!**


```python
plt.figure(figsize=[18,6])
base_color = sns.color_palette()[0]

plt.subplot(1,3,1)
cut_order = ['Fair', 'Good', 'Very Good', 'Premium']
sns.countplot(data=df, x='cut', color=base_color, order=cut_order)

plt.subplot(1,3,2)
color_order = ['J', 'I', 'H', 'G', 'F', 'E', 'D']
sns.countplot(data=df, x='color', color=base_color, order=color_order)

plt.subplot(1,3,3)
clarity_order = ['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF']
sns.countplot(data=df, x='clarity', color=base_color, order=clarity_order)
```

```python
# convert cut, color, and clarity into ordered categorical types
ordinal_var_dict = {'cut': ['Fair','Good','Very Good','Premium','Ideal'],
                    'color': ['J', 'I', 'H', 'G', 'F', 'E', 'D'],
                    'clarity': ['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF']}

for var in ordinal_var_dict:
    pd_ver = pd.__version__.split(".")
    if (int(pd_ver[0]) > 0) or (int(pd_ver[1]) >= 21): # v0.21 or later
        ordered_var = pd.api.types.CategoricalDtype(ordered = True,
                                                    categories = ordinal_var_dict[var])
        df[var] = df[var].astype(ordered_var)
    else: # pre-v0.21
        df[var] = df[var].astype('category', ordered = True,
                                             categories = ordinal_var_dict[var])
```

## Bivariate Exploration
In the previous notebook, you looked at the univariate distribution of five features in the diamonds dataset: price, carat, cut, color, and clarity. Now, we'll investigate relationships between pairs of these variables, particularly how each of them relate to diamond price.

To start, construct a plot of the price against carat weight. What kind of shape does the relationship between these variables take?

```python
plt.figure(figsize=[10,6])
sns.regplot(data=df, x='carat', y='price', fit_reg=False, scatter_kws = {'alpha' : 1/10})
```

> We can see that the graph exhibits a concave upward shape.

```python
# applying a log transform since the price distribution has a very long tail
g = sns.FacetGrid(data=df, height=8, aspect=1.2)
g.map(sns.regplot,'carat', 'price', fit_reg=False, scatter_kws = {'alpha' : 1/12})
g.set(yscale='log')
y_ticks = [300, 1000, 3000, 10000]
g.set(yticks = y_ticks, yticklabels = y_ticks)
g.add_legend()
```

> We can see that the graph exhibits a concave downward shape

```python
# applying a log transform on price and a cube transform on weight
df['cube_root_carat'] = df['carat'] ** (1/3)
g = sns.FacetGrid(data=df, height=8, aspect=1.2)
g.map(sns.regplot,'cube_root_carat', 'price', fit_reg=False, scatter_kws = {'alpha' : 1/12})
g.set(yscale='log')
y_ticks = [300, 1000, 3000, 10000]
g.set(yticks = y_ticks, yticklabels = y_ticks)
g.add_legend()
```

> Now the graph looks linear!

Now let's take a look at the relationship between price and the three categorical quality features, cut, color, and clarity. Are there any surprising trends to be seen here?

```python
plt.figure(figsize=[18,6])
base_color = sns.color_palette()[0]

plt.subplot(1,2,1)
cut_order = ['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']
sns.boxplot(data=df, x='cut', y='price', color=base_color, order=cut_order)

plt.subplot(1,2,2)
cut_order = ['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']
sns.violinplot(data=df, x='cut', y='price', color=sns.color_palette()[1], order=cut_order)
```

```python
plt.figure(figsize=[18,6])

plt.subplot(1,2,1)
color_order = ['J', 'I', 'H', 'G', 'F', 'E', 'D']
sns.boxplot(data=df, x='color', y='price', color=base_color, order=color_order)

plt.subplot(1,2,2)
color_order = ['J', 'I', 'H', 'G', 'F', 'E', 'D']
sns.violinplot(data=df, x='color', y='price', color=sns.color_palette()[1], order=color_order)
```

```python
plt.figure(figsize=[18,6])

plt.subplot(1,2,1)
clarity_order = ['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF']
sns.boxplot(data=df, x='clarity', y='price', color=base_color, order=clarity_order)

plt.subplot(1,2,2)
clarity_order = ['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF']
sns.violinplot(data=df, x='clarity', y='price', color=sns.color_palette()[1], order=clarity_order)
```

## Multivariate Exploration

In the previous workspace, you looked at various bivariate relationships. You saw that the log of price was approximately linearly related to the cube root of carat weight, as analogy to its length, width, and depth. You also saw that there was an unintuitive relationship between price and the categorical quality measures of cut, color, and clarity, that the median price decreased with increasing quality. Investigating the distributions more clearly and looking at the relationship between carat weight with the three categorical variables showed that this was due to carat size tending to be smaller for the diamonds with higher categorical grades.

The goal of this workspace will be to depict these interaction effects through the use of multivariate plots.

To start off with, create a plot of the relationship between price, carat, and clarity. In the previous workspace, you saw that clarity had the clearest interactions with price and carat. How clearly does this show up in a multivariate visualization?

```python
g = sns.FacetGrid(data=df, col='clarity', size=4, col_wrap=4)
g.map(sns.regplot, 'carat', 'price', scatter_kws={'alpha':1/3}, line_kws={"color": "red"})
```

```python
g = sns.FacetGrid(data=df, hue='clarity', height = 9, aspect = 1.5, palette = 'viridis')
g.map(sns.regplot, 'carat', 'price', fit_reg = False);
g.add_legend();
```

```python
sns.distplot(df['price'], kde=False)
```

> Recall that the price and weight are highly skewed. Therefore, before plotting, we should **transform** both axises.

```python
g = sns.FacetGrid(data=df, col='clarity', size=4, col_wrap=4)
g.map(sns.regplot, 'cube_root_carat', 'price', fit_reg=False, scatter_kws={'alpha':1/3}, line_kws={"color": "red"})
g.set(yscale='log')
y_ticks = [300, 1000, 3000, 10000]
g.set(yticks = y_ticks, yticklabels = y_ticks)
```
